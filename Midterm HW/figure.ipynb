```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_image(image):
    plt.imshow(image, cmap='binary', interpolation='nearest')
    plt.xticks(np.arange(len(image)))
    plt.yticks(np.arange(len(image[0])))
```

```python
def psi(image):
    height = len(image)
    width = len(image[0])
    new_image = []
    for i in range(0, height):
        row = []
        for j in range(0, width):
            u2_x = 2*i
            u2_y = 2*j
            cnt = 0
            addSum = 0
            for x in [-1, 0, 1]:
                for y in [-1, 0, 1]:
                    v_x = u2_x + x
                    v_y = u2_y + y
                    if v_x >=0 and v_x < height and v_y >=0 and v_y < width:
                        cnt+=1
                        addSum+=image[v_x][v_y]
            #print()
            if cnt > 0:
                row.append(addSum/cnt)
            # else:
            #     row.append(0)
        if row:
            new_image.append(row)
    return new_image
```
